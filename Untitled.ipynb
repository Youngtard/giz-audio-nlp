```python
# Ref: https://www.kaggle.com/haqishen/augmentation-methods-for-audio
def get_spectrogram(wav, log = True):
    # short-time Fourier transform
    D = librosa.stft(wav, n_fft=480, hop_length=160,
                     win_length=480, window='hamming')
    # Default args
#     D = librosa.stft(wav,)
    spect, phase = librosa.magphase(D)
    if log:
        # Log transform
#         log_spect = np.log(spect)
        # By converting to decibels (dB) the scale becomes logarithmic
        # https://stackoverflow.com/questions/63347977/what-is-the-conceptual-purpose-of-librosa-amplitude-to-db
        log_spect = librosa.amplitude_to_db(spect, ref=np.max)
        return log_spect
    return spect
    #plt.savefig(f'img_data/{g}/{filename[:-3].replace(".", "")}.png')
```

```python
#https://github.com/TheoViel/kaggle_birdcall_identification/blob/master/src/data/transforms.py
def mono_to_color(X, eps=1e-6, mean=None, std=None):
    """
    Converts a one channel array to a 3 channel one in [0, 255]
    Arguments:
        X {numpy array [H x W]} -- 2D array to convert
    Keyword Arguments:
        eps {float} -- To avoid dividing by 0 (default: {1e-6})
        mean {None or np array} -- Mean for normalization (default: {None})
        std {None or np array} -- Std for normalization (default: {None})
    Returns:
        numpy array [3 x H x W] -- RGB numpy array
    """
    X = np.stack([X, X, X], axis=-1)

    # Standardize
    mean = mean or X.mean()
    std = std or X.std()
    X = (X - mean) / (std + eps)

    # Normalize to [0, 255]
    _min, _max = X.min(), X.max()

    if (_max - _min) > eps:
        V = np.clip(X, _min, _max)
        V = 255 * (V - _min) / (_max - _min)
        V = V.astype(np.uint8)
    else:
        V = np.zeros_like(X, dtype=np.uint8)

    return V
```

```python
# def get_melspectogram(wav, cst=5, top_db=80):
def get_melspectogram(wav):    
    mel_spect = librosa.feature.melspectrogram(wav)
    # Convert to log scale (dB). We'll use the peak power (max) as reference.
    # https://www.kaggle.com/davids1992/speech-representation-and-data-exploration
    mel_spect = librosa.power_to_db(mel_spect)
    
    s = mono_to_color(mel_spect)

    return s
```

```python
def plot_spectrogram(spect):
    print('spectrogram shape:', spect.shape)
    plt.imshow(spect, aspect='auto', origin='lower',)
    plt.title('spectrogram of origin audio')
    plt.show()
```
